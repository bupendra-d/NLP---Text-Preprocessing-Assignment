```python
import nltk
import string

from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
# Input paragraphs
text = "Artificial Intelligence (AI) is transforming the modern world. Many industries, such as healthcare, finance, and education, are adopting AI technologies. However, people often ask: will machines replace humans? Experts say AI will assist humans rather than replace them completely! Therefore, learning AI skills is becoming important for students."
"Natural Language Processing (NLP) helps computers understand human language. It is used in applications like chatbots, search engines, and voice assistants. But raw text data usually contains punctuation marks, stop words, and other unnecessary information; therefore, preprocessing is required. Cleaning the data improves the performance of machine learning models!"

print(text)
```

```python
# Tokenization
tokens = word_tokenize(text)
```

```python
# 1. Remove punctuation
tokens_no_punctuation = [word for word in tokens if word not in string.punctuation]
print("\nAfter Removing Punctuation:")
print(tokens_no_punctuation)
```

```python
# 2. Stopword Removal
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens_no_punctuation if word.lower() not in stop_words]

print("\nAfter Stopword Removal:")
print(filtered_words)
```

```python
# 3. Stemming
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_words]

print("\nAfter Stemming:")
print(stemmed_words)
```

```python
# 4. Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]

print("\nAfter Lemmatization:")
print(lemmatized_words)
```
